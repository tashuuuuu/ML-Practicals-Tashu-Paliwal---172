```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
```

```python

iris = load_iris()

df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)


df.head()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
# Set plot style
sns.set(style="whitegrid")

# Plot boxplots for each feature
plt.figure(figsize=(14, 10))
for i, column in enumerate(iris.feature_names):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(x='species', y=column, data=df)
    plt.title(f'Box plot of {column}')
    plt.xlabel('Species')
    plt.ylabel(column)

plt.tight_layout()
plt.show()
```

```python
# Set plot style
sns.set(style="whitegrid")

# Plot histograms
df.hist(figsize=(12, 8), bins=20, color='skyblue', edgecolor='black')
plt.suptitle("Histograms of Iris Features", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Plot colored histograms for each feature
features = iris.feature_names
plt.figure(figsize=(16, 12))

for i, feature in enumerate(features):
    plt.subplot(2, 2, i + 1)
    sns.histplot(data=df, x=feature, hue='species', kde=True, palette='Set1', bins=20, alpha=0.6)
    plt.title(f'{feature.title()} Distribution by Species')
    plt.xlabel(feature.title())
    plt.ylabel("Count")

plt.tight_layout()
plt.show()
```

```python
# Calculate correlation matrix (numeric features only)
corr = df.iloc[:, :-1].corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Correlation Heatmap of Iris Features")
plt.show()
```
